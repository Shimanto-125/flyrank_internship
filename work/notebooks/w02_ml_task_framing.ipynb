## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

###My lane as an ML task is: Refresh / Content Opportunity Scoring:

**Task Type:** Classification

This is a classification problem because the goal is to predict whether a piece of content should be refreshed based on its performance trend. Each content item belongs to a category such as declining performance or not declining, allowing the model to support refresh decisions.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Answer:

The target is **trend_direction**.

It is an observed outcome derived from historical content performance.

Example classes:

- down
- stable
- up

The model will learn to predict the future performance trend of a content item using historical engagement, search, freshness, and traffic features.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

###Answer:

Primary Metric: Macro F1-score

The dataset contains multiple classes (up, stable, down). Macro F1-score measures performance across all classes equally and is suitable when class distributions may be imbalanced.

A good model should achieve a Macro F1-score higher than a simple baseline while maintaining balanced precision and recall.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

###Answer:

A simple rule such as "refresh every article older than 180 days" ignores many important signals.

Content performance depends on multiple interacting factors, including:

- Search volume
- Competition
- CTR
- Average position
- Engagement rate
- Traffic trend
- Content age
- Freshness
- AI traffic percentage

These relationships are too complex for a single if-statement. Machine learning can learn patterns from historical data and make better refresh recommendations than a fixed rule.
